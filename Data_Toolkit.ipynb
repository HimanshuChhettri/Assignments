```python
# (1) Demonstrate three different methods fpr creating identical 2D arrays in NumPy. Provides the code for each method and the final output after each method.

# Using np.array() with a list of lists:
import numpy as np

array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[1, 2, 3], [4, 5, 6]])

print("array1:", array1)
print("array2:", array2)

# Using np.tile() to repeat a 1D array:
import numpy as np

array1 = np.array([1, 2, 3])
array2 = np.tile(array1, (2, 1))

print("array1:", array1)
print("array2:", array2)

#  Using np.copy() to create a deep copy:
import numpy as np

array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = array1.copy()

print("array1:", array1)
print("array2:", array2)
```

```python
# (3) Explain the following terms:
# The differnce in np.asarray and ap.asanyarray
# The differnce between Deep copy and Shallow copy

# np.array(): Creates a new NumPy array from a Python sequence or array-like object.
# np.asarray(): Converts an array-like object to a NumPy array without copying the data if possible. If the input is already a NumPy array, it returns a reference to the original array.
# np.asanyarray(): Similar to np.asarray(), but it also allows for more flexibility in handling input types.

# Deep copy: Creates a new array with its own data, independent of the original array. Modifying one array does not affect the other.
# Shallow copy: Creates a new array that references the same data as the original array. Modifying one array affects the other.
```

```python
# (13) Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1. Perform the following tasks using Matplotlib and NumPy:
#    (a) Create a scatter plot using x and y, setting the color of the points to red and the marker style to 'o'.
#    (b) Add a horizontal line at y = 0.5 using a dashed line style and label it as 'y = 0.5'.
#    (c) Add a vertical line at x = 0.5 using a dotted line style and label it as 'x = 0.5'.
#    (d) Label the x-axis as 'X-axis' and the y-axis as 'Y-axis'.
#    (e) Set the title of the plot as 'Advanced Scatter Plot of Random Values'.
#    (f) Display a legend for the scatter plot, the horizontal line, and the vertical line.

import numpy as np
import matplotlib.pyplot as plt

# Create two NumPy arrays with random float values
x = np.random.rand(100)
y = np.random.rand(100)

# Create a scatter plot
plt.scatter(x, y, color='red', marker='o', label='Data Points')

# Add horizontal and vertical lines
plt.axhline(y=0.5, color='black', linestyle='--', label='y = 0.5')
plt.axvline(x=0.5, color='black', linestyle=':', label='x = 0.5')

# Label axes and set title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Advanced Scatter Plot of Random Values')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
# (14) Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
#    (a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
#    (b) Label the x-axis as 'Date'.
#    (c) Set the title of the plot as 'Temperature and Humidity Over Time'.

import pandas as pd
import matplotlib.pyplot as plt

# Create a sample time-series dataset
data = {
    'Date': pd.date_range(start='2023-01-01', end='2023-12-31'),
    'Temperature': np.random.randint(10, 35, 365),
    'Humidity': np.random.randint(50, 90, 365)
}

df = pd.DataFrame(data)

# Create a figure and axes
fig, ax1 = plt.subplots()

# Plot 'Temperature' on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='blue', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='blue')

# Create a second axis for 'Humidity' on the right y-axis
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='red', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='red')

# Set the title
plt.title('Temperature and Humidity Over Time')

# Show the legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# (15) Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
#    (a) Plot a histogram of the data with 30 bins.
#    (b) Overlay a line plot representing the normal distribution's probability density function (PDF).
#    (c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
#    (d) Set the title of the plot as 'Histogram with PDF Overlay'.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a histogram
plt.hist(data, bins=30, density=True, label='Histogram')

# Calculate the PDF of the normal distribution
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, loc=0, scale=1)

# Overlay the PDF as a line plot
plt.plot(x, pdf, color='red', label='PDF')

# Label axes and set title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
# (16) Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
x = np.random.randn(500)
y = np.random.randn(500)

# Determine quadrant for each point
quadrants = np.where(x >= 0, np.where(y >= 0, 1, 4), np.where(y >= 0, 2, 3))

# Create a Seaborn scatter plot with hue based on quadrants
sns.scatterplot(x=x, y=y, hue=quadrants, palette=['green', 'blue', 'orange', 'red'])

# Add legend
plt.legend(title='Quadrant')

# Label axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set title
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

```python
# (17) With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

import numpy as np
from bokeh.plotting import figure, show

# Generate x and y values for the sine wave
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add a line plot of the sine wave
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

```python
# (18) Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 30, 5)

# Create a ColumnDataSource for the data
source = ColumnDataSource(data={'categories': categories, 'values': values})

# Create a figure
p = figure(title='Random Categorical Bar Chart', x_axis_label='Category', y_axis_label='Value')

# Create a bar chart
p.vbar(x='categories', top='values', width=0.5, source=source, color='values', palette='Viridis')

# Add a hover tool to display exact values
hover = HoverTool(tooltips=[('Category', '@categories'), ('Value', '@values')])
p.add_tools(hover)

# Show the plot
show(p)
```

```python
# (20) Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

import plotly.graph_objects as go
import numpy as np

# Generate random data
labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 30, 5)

# Create a figure
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=0.3)])

# Set the title
fig.update_layout(title='Interactive Pie Chart')

# Show the plot
fig.show()
```

```python
# (19) Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.

import plotly.graph_objects as go
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)
y = np.random.randn(100)

# Create a figure
fig = go.Figure()

# Add a trace to the figure
fig.add_trace(go.Scatter(x=x, y=y, mode='lines'))

# Set the title and labels
fig.update_layout(title='Simple Line Plot', xaxis_title='X-axis', yaxis_title='Y-axis')

# Show the plot
fig.show()
```
