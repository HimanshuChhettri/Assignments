```python
# 1) Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

x = df['X']
y = df['Y']
z = df['Z']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.title('3D Scatter Plot')

plt.show()
```

```python
# 4) Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(sqrt x^2 + y^2).

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(y, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap='viridis', alpha=0.7)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

plt.show()
```

```python
# 5) Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(25)

data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

plt.scatter(df['GDP'], df['Population'], s=df['Population'] * 10, alpha=0.7)
plt.title('Population vs. GDP')
plt.xlabel('GDP')
plt.ylabel('Population')
plt.grid(True)

for i, row in df.iterrows():
    plt.text(row['GDP'], row['Population'], row['Country'], ha='center', va='bottom')

plt.show()
```

```python
# 1) Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Set output to notebook for inline visualization
output_notebook()

# Create data
x = np.linspace(0, 10, 1000)
y = np.sin(x)

# Create a figure and line plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

```python
# 2) Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data
N = 100
x = np.random.rand(N) * 100
y = np.random.rand(N) * 100
sizes = np.random.randint(10, 100, N)
colors = np.random.rand(N, 3)  # Random RGB colors

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a figure
p = figure(title="Scatter Plot with Varying Sizes and Colors")

# Add scatter plot
p.scatter('x', 'y', size='sizes', color='colors', source=source)

# Show the plot
show(p)
```

```python
# 3) Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data={'fruits': fruits, 'counts': counts})

# Create a figure
p = figure(title="Fruit Counts", x_axis_label="Fruits", y_axis_label="Counts")

# Add a bar chart
p.vbar(x='fruits', top='counts', width=0.5, source=source)

# Show the plot
show(p)
```

```python
# 4) Create a Bokeh histogram to visualize the distribution of the given data.

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource
source = ColumnDataSource(data={'hist': hist, 'edges': edges[:-1]})

# Create a figure
p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")

# Add a histogram
p.quad(left='edges', right='edges', bottom=0, top='hist', source=source)

# Show the plot
show(p)
```

```python
# 5) Create a Bokeh heatmap using the provided dataset.

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper, HoverTool # Import LinearColorMapper
import numpy as np

# Generate random data
data_heatmap = np.random.rand(10, 10)

# Create x and y arrays
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create meshgrid
xx, yy = np.meshgrid(x, y)

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': xx.ravel(), 'y': yy.ravel(), 'z': data_heatmap.ravel()})

# Create a figure
p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y")

# Add a heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, fill_color='color', source=source)

# Set color mapper
color_mapper = LinearColorMapper(palette= 'Viridis256', low=data_heatmap.min(), high=data_heatmap.max())
p.add_tools (HoverTool(tooltips=[("x", "@x"), ("y", "@y"), ("value", "@z")]))
p.rect (x="x", y="y", width=0.1, height=0.1, source=source, fill_color={'field': 'z', 'transform': color_mapper})

# Show the plot
show(p)
```
